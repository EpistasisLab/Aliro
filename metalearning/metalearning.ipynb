```python
import pandas as pd
import numpy as np

# get data metafeatures
metafeatures = pd.read_csv('data_metafeatures.csv',sep=',',index_col=0)
print('loaded ', metafeatures.shape[1]-1, ' metafeatures for ', metafeatures.shape[0], ' datasets')
# get ML results
data = pd.read_csv('sklearn-benchmark5-data-edited.tsv.gz', sep='\t', names=['dataset',
                                                                     'classifier',
                                                                     'parameters',
                                                                     'accuracy', 
                                                                     'macrof1',
                                                                     'bal_accuracy']).fillna('')

data['accuracy'] = data['accuracy'].apply(lambda x: round(x, 3))
print('loaded ',data['dataset'].unique().shape[0],'datasets and ', data['classifier'].unique().shape[0],'classifiers')
# subset data to classifiers used in PennAI
pennai_classifiers = ['LogisticRegression', 'RandomForestClassifier', 'SVC', 
                      'KNeighborsClassifier', 'DecisionTreeClassifier', 'GradientBoostingClassifier']
mask = np.array([c in pennai_classifiers for c in data['classifier'].values])
data = data.loc[mask,:]
print('classifiers:',data['classifier'].unique())

print(data[:10])
```

### OPTIONAL: Preparing a dataset combining PMLB dataset with ML+P

```python
#set proper indexes
metafeatures=metafeatures.set_index('dataset')
data=data.set_index('dataset')
#see what were the methods
method_names=data['classifier'].unique().tolist()
method_names

#prepare new data file 
data2=pd.DataFrame()

for name_id in range(0,len(method_names)):
    name=pd.DataFrame()
    #check for the parameters of ML methods
    z=data[data.classifier==method_names[name_id]]['parameters'].str.split(',', expand=True)
    for i in range(0,len(z.columns)):
        value=z[i].str.find("=")
        name[i]=z[i].str[:value[0]]
        z[i]=z[i].str[value[0]+1:]
    z.columns=method_names[name_id]+"."+name.iloc[0]
    data2=data2.append(z)
df=pd.concat([metafeatures, data2], axis=1, join="inner")


#store data to file 
df.to_csv("metafeatures-nn.csv",sep="\t",)
```
